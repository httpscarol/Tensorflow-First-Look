# **1 - Importing and Sorting Data**

```python
import tensorflow
import pandas as pds
from sklearn.model_selection import train_test_split

df = pds.read_csv('https://raw.githubusercontent.com/httpscarol/Tensorflow-First-Look/main/Churn.csv')

df
```

```python
total = pds.get_dummies(df.drop(['Churn', 'Customer ID'], axis=1))

classify = df['Churn'].apply(lambda x:1 if x=='Yes' else 0)
```

```python
total_treino, total_teste, classify_treino, classify_teste = train_test_split(total, classify, test_size=.2)
```
